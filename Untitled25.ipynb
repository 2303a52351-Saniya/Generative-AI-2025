```python
import numpy as np

# Sigmoid and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Training and testing data
X_train = np.array([
    [0.2, 0.1],
    [0.3, 0.2],
    [0.4, 0.3],
    [0.7, 0.6],
    [0.8, 0.7],
    [0.9, 0.8]
])
y_train = np.array([[0.3441], [0.3500], [0.3558], [0.3729], [0.3785], [0.3841]])

X_test = np.array([
    [0.5, 0.4],
    [0.6, 0.5]
])
y_test = np.array([[0.3615], [0.3672]])

# Parameters
learning_rate = 0.1
epochs = 1000
input_neurons = 2
hidden_neurons = 4  # Hidden layer with 4 neurons
output_neurons = 1

# Initialize weights and biases
np.random.seed(42)
W_input_hidden = np.random.randn(input_neurons, hidden_neurons)
b_hidden = np.random.randn(hidden_neurons)

W_hidden_output = np.random.randn(hidden_neurons, output_neurons)
b_output = np.random.randn(output_neurons)

# Training loop
for epoch in range(epochs):
    # Forward pass
    hidden_input = np.dot(X_train, W_input_hidden) + b_hidden
    hidden_output = sigmoid(hidden_input)

    final_input = np.dot(hidden_output, W_hidden_output) + b_output
    final_output = final_input  # Linear activation in the output layer

    # Mean square error
    error = final_output - y_train
    mse = np.mean(error**2)

    # Backpropagation
    # Output layer gradients
    d_output = 2 * error / len(X_train)  # Gradient of MSE wrt output
    dW_hidden_output = np.dot(hidden_output.T, d_output)
    db_output = np.sum(d_output, axis=0)

    # Hidden layer gradients
    d_hidden = np.dot(d_output, W_hidden_output.T) * sigmoid_derivative(hidden_input)
    dW_input_hidden = np.dot(X_train.T, d_hidden)
    db_hidden = np.sum(d_hidden, axis=0)

    # Update weights and biases
    W_hidden_output -= learning_rate * dW_hidden_output
    b_output -= learning_rate * db_output
    W_input_hidden -= learning_rate * dW_input_hidden
    b_hidden -= learning_rate * db_hidden

    if epoch % 100 == 0:
        print(f"Epoch {epoch}, MSE: {mse:.5f}")

# Evaluate the model on test data
hidden_test_input = np.dot(X_test, W_input_hidden) + b_hidden
hidden_test_output = sigmoid(hidden_test_input)

final_test_input = np.dot(hidden_test_output, W_hidden_output) + b_output
test_predictions = final_test_input  # Linear activation

# Calculate test MSE
test_error = test_predictions - y_test
test_mse = np.mean(test_error**2)

print("\nTraining MSE:", mse)
print("Testing MSE:", test_mse)

# User input prediction
x_input = np.array([float(i) for i in input("\nEnter x1 and x2 separated by spaces: ").split()]).reshape(1, -1)

# Forward pass for user input
hidden_input_user = np.dot(x_input, W_input_hidden) + b_hidden
hidden_output_user = sigmoid(hidden_input_user)

final_input_user = np.dot(hidden_output_user, W_hidden_output) + b_output
user_prediction = final_input_user  # Linear output

print("\nPredicted output:", user_prediction[0][0])
```
